## Same function as in the speclib generation

```python
def filter_prediction(csv_file, PTM, threshold, control=False):
    """
    Filters predictions from a CSV file based on a threshold and adds 'residue' and 'label' fields.
    If control is True, returns random entries instead of those filtered by the threshold.
    
    Parameters:
    - csv_file (str): Path to the input CSV file.
    - PTM (str): The PTM type to extract the residue and label information.
    - threshold (float): The threshold value for filtering the 'pred' column.
    - control (bool): If True, returns random entries instead of filtered ones.
    
    Returns:
    - List[dict]: A list of dictionaries with keys 'prot_id', 'position', 'pred', 'residue', and 'label'.
    """
    # Read the CSV file
    df = pd.read_csv(csv_file)
    original_row_count = len(df)
    
    # Filter the DataFrame based on the threshold
    filtered_df = df[df['pred'] >= threshold].copy()
    filtered_row_count = len(filtered_df)
    
    # Parse the PTM to get 'residue' and 'label'
    if '[' in PTM and ']' in PTM:
        residue, label = PTM.split('[')
        label = label.replace(']', '')
    else:
        raise ValueError(f"Invalid PTM format: {PTM}. Expected format 'residue[label]'.")
    
    # Create the list of dictionaries to return
    result = []
    
    if control:
        # If control is True, randomly sample from the entire dataframe
        random_sample = df.sample(n=filtered_row_count, random_state=42)  # For reproducibility, random_state can be set
        for _, row in random_sample.iterrows():
            entry = {
                'prot_id': row['prot_id'],
                'position': row['position'],
                'pred': row['pred'],
                'residue': residue,
                'label': label
            }
            result.append(entry)
        print(f"Control Library: {filtered_row_count} random entries sampled from {original_row_count} rows.")
    else:
        # Regular filtering and return
        for _, row in filtered_df.iterrows():
            entry = {
                'prot_id': row['prot_id'],
                'position': row['position'],
                'pred': row['pred'],
                'residue': residue,
                'label': label
            }
            result.append(entry)
        print(f"{filtered_row_count} / {original_row_count} rows retained after filtering for {PTM}")
    
    return result
```

```python
def plot_prediction_distributions_hist(target_ptm_list, thresholds=[0.00, 0.05, 0.1, 0.2, 0.4, 0.8, 0.99]):
    """
    Creates a histogram-based plot showing the distribution of prediction values on a log scale.
    
    Args:
        target_ptm_list: List of lists containing dictionaries with prediction information
        thresholds: List of threshold values used for filtering
    """
    # Set style for scientific publication
    plt.style.use('seaborn-v0_8-whitegrid')
    
    # Create figure
    fig, ax = plt.subplots(figsize=(6, 6))
    
    # Define bins and custom colors
    bins = np.linspace(0, 1, 101)  # 101 bins for precise visualization
    custom_colors = ['#E41A1C', '#377EB8', '#4DAF4A', '#984EA3', '#FF7F00', '#FFFF33', 'Grey']
    
    # Plot histograms for each threshold
    max_count = 0
    for target_ptms, threshold, color in zip(target_ptm_list, thresholds, custom_colors):
        # Extract prediction values
        pred_values = [item['pred'] for item in target_ptms]
        
        # Calculate histogram
        counts, edges = np.histogram(pred_values, bins=bins)
        max_count = max(max_count, max(counts))
        
        # Plot histogram
        ax.hist(pred_values, bins=bins, alpha=0.7, color=color,
               label=f'Threshold {threshold:.2f} (n={len(pred_values):,})')

    # Set log scale
    ax.set_yscale('log')
    
    # Customize plot
    ax.set_xlabel('Prediction Score', fontsize=14)
    ax.set_ylabel('Phosphosite Count (log scale)', fontsize=14)
    
    # Set y-limits
    ax.set_ylim(1, max_count * 2)
    
    # Add threshold markers and set x-ticks with rotation
    ax.set_xticks(thresholds)
    ax.set_xticklabels([f'{t:.2f}' for t in thresholds], rotation=45, ha='right', fontsize=14)  # Increased tick label size
    
    # Increase y-axis tick label size
    ax.tick_params(axis='y', labelsize=14)
    
    for threshold in thresholds:
        if threshold > 0:  # Skip the 0 threshold
            ax.axvline(x=threshold, color='gray', linestyle='--', alpha=0.5)
    
    # Customize grid
    ax.grid(True, alpha=0.3)
    
    # Add legend with box
    legend = ax.legend(bbox_to_anchor=(0.98, 0.98), loc='upper right',
                      frameon=True, fancybox=False, framealpha=1.0,
                      edgecolor='black', facecolor='white',
                      bbox_transform=ax.transAxes)
    
    # Adjust layout to prevent label cutoff
    plt.tight_layout()
    
    return fig, ax

# Use the function
fig, axes = plot_prediction_distributions_hist(target_ptm_xxx)
plt.show()
```

```python
# Regular: 
Regular = [15921219, 9912333, 8863342, 7893360, 7044506, 6221585, 5837799]
Phospho_Enriched = [49721965, 20095411, 15298342, 10984245, 7664024, 5016302, 4230916]


def plot_library_size_reduction(regular, phospho_enriched, thresholds=[0.00, 0.05, 0.1, 0.2, 0.4, 0.8, 0.99]):
    """
    Creates a line plot showing the reduction in library size across different filtering thresholds.
    
    """
    # Set style for scientific publication
    plt.style.use('seaborn-v0_8-whitegrid')
    
    # Create figure
    fig, ax = plt.subplots(figsize=(6, 6))
    
    # Plot lines with different colors and markers
    ax.plot(thresholds, regular, 'o-', color='#2563eb', linewidth=2, 
            label='Regular Library', markersize=8, markeredgecolor='white', markeredgewidth=1)
    ax.plot(thresholds, phospho_enriched, 's-', color='#dc2626', linewidth=2, 
            label='Phospho-enriched Library', markersize=8, markeredgecolor='white', markeredgewidth=1)
    
    # Set linear scale for y-axis with 5M increments
    ax.set_ylim(0, 51000000)
    y_ticks = np.arange(0, 51000000, 5000000)
    ax.set_yticks(y_ticks)
    ax.set_yticklabels([f'{int(y/1000000)}M' for y in y_ticks], fontsize=14)
    
    # Increase size of tick labels
    ax.tick_params(axis='both', which='major', labelsize=14)
    
    # Make tick marks longer and thicker
    ax.tick_params(axis='both', which='major', width=1.5, length=6)
    
    # Set axis labels with improved spacing
    ax.set_xlabel('Prediction Score Threshold', fontsize=14, labelpad=10)
    ax.set_ylabel('Number of Precursors in Library', fontsize=14, labelpad=10)
    
    # Format x-axis
    ax.set_xticks(thresholds)
    ax.set_xticklabels([f'{t:.2f}' for t in thresholds], rotation=45, ha='center')
    
    # Customize grid
    ax.grid(True, linestyle='--', alpha=0.7)
    
    # Add legend inside the figure
    legend = ax.legend(loc='upper right', frameon=True, fancybox=False, 
                      framealpha=1.0, edgecolor='black', facecolor='white',
                      fontsize = 14)
    
    # Add a white background patch to the legend
    legend.get_frame().set_facecolor('white')
    legend.get_frame().set_linewidth(0.5)
    
    # Adjust layout to prevent label cutoff
    plt.tight_layout()
    
    return fig, ax

fig, ax = plot_library_size_reduction(regular=Regular, phospho_enriched=Phospho_Enriched)
plt.show()
```

```python
def plot_analysis_time_reduction(enriched_timstof_time, enriched_orbitrap_exploris_time, 
                               regular_orbitrap_time, regular_TTOFF_time, 
                               thresholds=[0.00, 0.05, 0.1, 0.2, 0.4, 0.8, 0.99]):
    """
    Creates a line plot showing the percentage reduction in analysis time across different filtering thresholds,
    relative to the initial time at threshold 0.00.
    """
    # Convert to percentage of initial value for each instrument
    timstof_pct = [100 * val / enriched_timstof_time[0] for val in enriched_timstof_time]
    exploris_pct = [100 * val / enriched_orbitrap_exploris_time[0] for val in enriched_orbitrap_exploris_time]
    orbitrap_pct = [100 * val / regular_orbitrap_time[0] for val in regular_orbitrap_time]
    ttof_pct = [100 * val / regular_TTOFF_time[0] for val in regular_TTOFF_time]
    
    # Set style for scientific publication
    plt.style.use('seaborn-v0_8-whitegrid')
    
    # Create figure
    fig, ax = plt.subplots(figsize=(6, 6))
    
    # Plot lines with different colors and markers
    ax.plot(thresholds, timstof_pct, 's-', color='#FD7676', linewidth=2, 
            label='Enriched timsTOF Pasef', markersize=8, markeredgecolor='white', markeredgewidth=1)
    ax.plot(thresholds, exploris_pct, '^-', color='#C40303', linewidth=2, 
            label='Enriched Orbitrap Exploris 480', markersize=8, markeredgecolor='white', markeredgewidth=1)
    ax.plot(thresholds, orbitrap_pct, 'o-', color='#013575', linewidth=2, 
            label='Regular Orbitrap AIF', markersize=8, markeredgecolor='white', markeredgewidth=1)
    ax.plot(thresholds, ttof_pct, 'D-', color='#76B3FD', linewidth=2, 
            label='Regular TTOF6600 SWATH', markersize=8, markeredgecolor='white', markeredgewidth=1)
    
    # Set linear scale for y-axis with 10% increments
    ax.set_ylim(0, 109)
    y_ticks = np.arange(0, 101, 10)
    ax.set_yticks(y_ticks)
    ax.set_yticklabels([f'{int(y)}%' for y in y_ticks], fontsize=12)
    
    # Increase size of tick labels
    ax.tick_params(axis='both', which='major', labelsize=12)
    
    # Make tick marks longer and thicker
    ax.tick_params(axis='both', which='major', width=1.5, length=6)
    
    # Set axis labels with improved spacing
    ax.set_xlabel('Prediction Score Threshold', fontsize=14, labelpad=10)
    ax.set_ylabel('Relative Analysis Time', fontsize=14, labelpad=10)
    
    # Format x-axis
    ax.set_xticks(thresholds)
    ax.set_xticklabels([f'{t:.2f}' for t in thresholds], rotation=45, ha='right')
    
    # Customize grid
    ax.grid(True, linestyle='--', alpha=0.7)
    
    # Add legend inside the figure
    legend = ax.legend(loc='upper right', frameon=True, fancybox=False, 
                      framealpha=1.0, edgecolor='black', facecolor='white',
                      fontsize=12)
    
    # Add a white background patch to the legend
    legend.get_frame().set_facecolor('white')
    legend.get_frame().set_linewidth(0.5)
    
    # Adjust layout to prevent label cutoff
    plt.tight_layout()
    
    return fig, ax
```

```python
# analysis times 
enriched_timstof_time = [1247, 441, 312, 229, 148, 102, 89]
enriched_orbitrap_exploris_time = [365, 143, 111, 80, 57, 45, 39] 
regular_orbitrap_time = [66, 41, 34, 30, 29, 26, 34]
regular_TTOFF_time = [367, 239, 207, 221, 191, 166, 164]
```
